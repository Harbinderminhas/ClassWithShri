# Iterating over a DataFrame

```python
# method 1: iteritems() returns a column in each iteration
df = pd.DataFrame(np.random.rand(5,4),  # an array of 5 rows and 4 columns
                  columns=['col1', 'col2', 'col3', 'col4'])
df
```

```python
for column_name, col_value in df.iteritems():
    print("column name is:", column_name)
    print(col_value)
    print("*"*50)
```

## df.iterrows(): Iterate row wise

```python
for row_index, row_value in df.iterrows():
    print("Row index:", row_index)
    print(row_value)
    print("*"*50)
```

## itertuples(): returns a named tuple`m

```python
for row in df.itertuples():
    print(row)
    print("*"*50)
```

```python
for row in df.itertuples():
    print(row.col1)
    print("*"*50)
```

# Merging and Joining

```python
import pandas as pd
```

```python
# inner join returns intersection of two Dataframes.
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[2,5,9],
           'Points':[874,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
df1
```

```python
print("DF2: Chokers")
df2
```

```python
print(" *******Default is Inner Join Merge*********")
pd.merge(df1, df2, on='Team')
# equivalent code
#pd.merge(df1, df2, on='Team', how='inner')
```

```python
# left join
pd.merge(df1, df2, on='Team', how='left')
```

```python
# right join
pd.merge(df1, df2, on='Team', how='right')
```

```python
# full outer join
pd.merge(df1, df2, on='Team', how='outer')
```
