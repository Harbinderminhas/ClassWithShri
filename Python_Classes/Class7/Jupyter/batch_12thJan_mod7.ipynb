## iterating over a dataframe

```python
# method 1: iteritems() returns a column in each iteration
df = pd.DataFrame(np.random.rand(5,4),  # an array of 5 rows and 4 columns
                  columns=['col1', 'col2', 'col3', 'col4'])
df
```

```python
for key, value in df.iteritems():
    print("KEY: ", key)
    print(value)
    print("#"*20)
```

```python
# method 2: iterrows()
df = pd.DataFrame(np.random.rand(5,4),  # an array of 5 rows and 4 columns
                  columns=['col1', 'col2', 'col3', 'col4'])
df
```

```python
for key, value in df.iterrows():
    print("KEY: ", key)
    print(value)
    print("#"*20)
```

```python
# method 3: itertuples()
df = pd.DataFrame(np.random.rand(5,4),  # an array of 5 rows and 4 columns
                  columns=['col1', 'col2', 'col3', 'col4'])
df
```

```python
for row in df.itertuples():
    print(row)
    #print(row.col1)
    print("#"*20)
```

# GroupBy operations on DataFrames

```python
import pandas as pd

world_cup={'Team':['West Indies','West Indies','India','Australia','Pakistan','Sri Lanka','Australia','Australia','Australia','India','Australia'],
           'Rank':[7,7,2,1,6,4,1,1,1,2,1],
           'Year':[1975,1979,1983,1987,1992,1996,1999,2003,2007,2011,2015]}
df=pd.DataFrame(world_cup)
print(df)
```

```python
df.groupby('Team')
```

```python
df.groupby('Team').groups
```

```python
# iterating through groups
grouped = df.groupby('Team')
for name, value in grouped:
    print(name)
```

```python
# selecting a group
grouped = df.groupby('Team')
grouped.get_group('India')
```

```python
grouped.get_group('Australia')
```

### concatenation: appending one dataframe to another one

```python
# method 1: using pandas.concat()
import pandas as pd

world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['South Africa','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[895,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
```

```python
new_df = pd.concat([df1,df2],sort=True)
new_df
```

```python
new_df.loc[0]  # label 0
```

```python
new_df.iloc[0] # 0th position
```

```python
# method 2: df.append()
new_df = df1.append(df2, sort=False)
new_df
```

## Merging and joining

```python
# inner join returns intersection of two Dataframes.
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[2,5,9],
           'Points':[874,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
print(" *******Default is Inner Join Merge*********")
print(pd.merge(df1, df2, on='Team'))
```

```python
# left join
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
print("######### LEFT JOIN ###########")
print(pd.merge(df1, df2, on='Team', how='left'))
```

```python
# right join
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
print("######### RIGHT JOIN ###########")
print(pd.merge(df1, df2, on='Team', how='right'))
```

```python
# outer join
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
print("######### OUTER JOIN ###########")
print(pd.merge(df1, df2, on='Team', how='outer'))
```

```python
# inner join is by default is you don't provide 'how'
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
print("######### inner JOIN ###########")
print(pd.merge(df1, df2, on='Team', how='inner'))
```

```python
# 1.list of countries having area greater than 2000
selected_data = dataset.loc[:,['Country', 'LandArea']]
selected_data.head()
```

```python
# 1.list of countries having area greater than 2000
selected_data[selected_data['LandArea'] > 2000]
```

```python
# multiple conditions
selected_data[(selected_data['LandArea'] > 2000) & (selected_data['LandArea'] < 3000)]
```
