# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
# Line plot

plt.plot([11, 12, 13, 14])  # only giving y-axis values
plt.show()  # dislays the plot
```

```python
# giving both x and y axis data
x = [1, 4, 9, 10]
plt.plot(x, [elem**2 for elem in x])
plt.show()
```

```python
# using an array
import numpy as np
x = np.arange(0,4,0.01)
plt.plot(x, [elem**2 for elem in x])
plt.show()
```

## Multiline plots

```python
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x])
plt.plot(x, [elem**2 for elem in x])
plt.plot(x, [elem**3 for elem in x])
plt.show()
```

### adding Grids

```python
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x])
plt.plot(x, [elem**2 for elem in x])
plt.plot(x, [elem**3 for elem in x])
plt.grid(True)
plt.show()
```

```python
# creating multiple plots with single API call
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
plt.show()
```

### Limiting the axes

```python
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
############
# Method 1
plt.axis([-1, 2, -1, 10])
###########
plt.show()
```

```python
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
############
plt.axis([1, 2, 1, 10])
###########
plt.show()
```

```python
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
############
# Method 2
plt.xlim(0, 2)
plt.ylim(0, 10)
###########
plt.show()
```

## Adding Labels

```python
# adding axes names
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
############
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
###########
plt.show()
```

```python
# adding a Title
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
###########
plt.title('Polynomial Graph')
###########
plt.show()
```

```python
# adding a legend
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x], label='Linear')
plt.plot(x, [elem**2 for elem in x], label='Square')
plt.plot(x, [elem**3 for elem in x], label='Cubic')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Graph')
###########
plt.legend()
######
plt.show()
```

### saving a plot

```python
x = np.arange(5)  # an array of 5 elements
plt.plot(x, x, label='Linear')
plt.plot(x, x**2, label='Square')
plt.plot(x, x**3, label='Cube')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Graph')
plt.legend()
######
plt.savefig('plot.png')
#####
plt.show()
```

## Bar Chart
- used to compare two or more values using rectangular bars

```python
plt.bar([1, 2, 3], [1, 4, 9])
plt.show()
```

```python
plt.bar([1, 2, 3, 5, 6], [1, 4, 9, 10, 20])
plt.show()
```

```python
# Method 1: for giving xticks
sales_data = {'bicyles':10, 'car':34, 'scooter':56, 'airplanes':4}
for elem in sales_data.items():
    plt.bar(elem[0], elem[1])
plt.show()
```

```python
# Method 2: for giving xticks
sales_data = {'bicyles':10, 'car':34, 'scooter':56, 'airplanes':4}
for i, key in enumerate(sales_data):
    print(i,key)
    plt.bar(i, sales_data[key])
plt.xticks([0,1,2,3], sales_data.keys())
plt.show()
```

## Pie chart

```python
sales_items = ['bicyles', 'car', 'scooter','airplanes']
sales_units = [10, 20, 34, 5]
# total items = 10+20+34+5 = 69
# % of bicyles = 10/69

plt.pie(sales_units, labels=sales_items)
plt.show()
```

```python
# display the percentage also
sales_items = ['bicyles', 'car', 'scooter','airplanes']
sales_units = [10, 20, 34, 5]
plt.pie(sales_units,
        labels=sales_items,
        autopct='%d%%')
# %d is for whole number
# %f is for float values
plt.show()
```

```python
# display the percentage in float values
sales_items = ['bicyles', 'car', 'scooter','airplanes']
sales_units = [10, 20, 34, 5]
plt.pie(sales_units,
        labels=sales_items,
        autopct='%.2f%%')
plt.show()
```

```python
# explode
sales_items = ['bicyles', 'car', 'scooter','airplanes']
sales_units = [10, 20, 34, 5]
plt.pie(sales_units,
        labels=sales_items,
        autopct='%.2f%%',
       explode=(0,0.2,0,0))
plt.show()
```

```python
# control the size of a pie chart
plt.figure(figsize=(3,3))
sales_items = ['bicyles', 'car', 'scooter','airplanes']
sales_units = [10, 20, 34, 5]

plt.pie(sales_units,
        labels=sales_items,
        autopct='%.2f%%')

plt.show()
```

```python
# control the size of a pie chart
plt.figure(figsize=(8,6))
sales_items = ['bicyles', 'car', 'scooter','airplanes']
sales_units = [10, 20, 34, 5]

plt.pie(sales_units,
        labels=sales_items,
        autopct='%.2f%%')

plt.show()
```

# Styling

## adding colors

```python
y = np.arange(1,3)
plt.plot(y, 'y')  # y is yellow
plt.plot(y+5, 'm')
plt.plot(y+10, 'c')
plt.show()
```

```python
y = np.arange(1,3)
plt.plot(y, 'g')  # y is yellow
plt.plot(y+5, 'b')
plt.plot(y+10, 'r')
plt.show()
```

### control line styling

```python
y = np.arange(1, 100)
plt.plot(y, '--', y*5, '-.', y*10,':')
plt.show()
```

### control marker styling

```python
y = np.arange(1, 3, 0.3)
plt.plot(y, '*', y+0.5, 'o', y+1,'D', y+2, '^', y+3, 's')
plt.show()
```
