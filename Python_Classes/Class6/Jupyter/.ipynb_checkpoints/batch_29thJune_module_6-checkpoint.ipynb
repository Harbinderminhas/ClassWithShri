# Data Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# line plot
plt.plot([1, 5, 10, 23, 56])  # value passed is the y-value
plt.show()
```

```python
# line plot using a Series of a DataFrame or Column

data = pd.read_csv('AllCountries.csv')
plt.plot(data['LandArea']) # by default the value passed is the y-value
plt.show()
```

```python
# line plot by providing both x and y values
x = data['GDP']
y = data['LandArea']
plt.plot(x,y)
plt.show()
```

```python
x = data['Country'][0:7]
y = data['GDP'][0:7]

plt.plot(x,y)
plt.show()
```

## Multiline plot

```python
# method 1
x = [1,2,3,4]
plt.plot(x, [elem for elem in x])
plt.plot(x, [elem**2 for elem in x])
plt.plot(x, [elem**3 for elem in x])
plt.show()
```

```python
# method 1
x = [1,2,3,4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.show()
```

```python
plt.plot(data['Country'][0:5], data['LandArea'][0:5],
         data['Country'][0:5], data['BirthRate'][0:5],
        data['Country'][0:5], data['DeathRate'][0:5],)
plt.show()
```

```python
# adding Grid
x = [1,2,3,4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True) ###############
plt.show()
```

```python
# limiting the axes
# method 1
x = [1,2,3,4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
plt.axis([0,2,0,10]) ###########
plt.show()
```

```python
# limiting the axes
# method 2
x = [1,2,3,4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
plt.xlim(0,3) ###########
plt.ylim(0,20)  #######
plt.show()
```
