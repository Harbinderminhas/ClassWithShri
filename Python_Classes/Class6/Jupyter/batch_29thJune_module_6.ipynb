# Data Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# line plot
plt.plot([1, 5, 10, 23, 56])  # value passed is the y-value
plt.show()
```

```python
# line plot using a Series of a DataFrame or Column

data = pd.read_csv('AllCountries.csv')
plt.plot(data['LandArea']) # by default the value passed is the y-value
plt.show()
```

```python
# line plot by providing both x and y values
x = data['GDP']
y = data['LandArea']
plt.plot(x,y)
plt.show()
```

```python
x = data['Country'][0:7]
y = data['GDP'][0:7]

plt.plot(x,y)
plt.show()
```

## Multiline plot

```python
# method 1
x = [1,2,3,4]
plt.plot(x, [elem for elem in x])
plt.plot(x, [elem**2 for elem in x])
plt.plot(x, [elem**3 for elem in x])
plt.show()
```

```python
# method 1
x = [1,2,3,4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.show()
```

```python
plt.plot(data['Country'][0:5], data['LandArea'][0:5],
         data['Country'][0:5], data['BirthRate'][0:5],
        data['Country'][0:5], data['DeathRate'][0:5],)
plt.show()
```

```python
# adding Grid
x = [1,2,3,4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True) ###############
plt.show()
```

```python
# limiting the axes
# method 1
x = [1,2,3,4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
plt.axis([0,2,0,10]) ###########
plt.show()
```

```python
# limiting the axes
# method 2
x = [1,2,3,4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
plt.xlim(0,3) ###########
plt.ylim(0,20)  #######
plt.show()
```

```python
# adding labels
x = [1,2,3,4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)

plt.xlabel('X-axis')  ########
plt.ylabel('Y-axis')  ########
plt.show()
```

```python
# adding title
x = [1,2,3,4]
plt.plot(x, [elem for elem in x],
        x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.title('Polynomial Graph')
plt.show()
```

```python
# adding Legend
x = [1,2,3,4]
plt.plot(x, [elem for elem in x], label='linear')
plt.plot(x, [elem**2 for elem in x], label='square')
plt.plot(x, [elem**3 for elem in x], label='cube')
plt.grid(True)
plt.legend() ############
plt.show()
```

```python
# saving a plot
x = [1,2,3,4]
plt.plot(x, [elem for elem in x], label='linear')
plt.plot(x, [elem**2 for elem in x], label='square')
plt.plot(x, [elem**3 for elem in x], label='cube')
plt.grid(True)
plt.legend()

plt.savefig('myplot_29thJune.pdf') #######
plt.show()
```

# Bar chart

```python
country = ['India', 'US', 'China', 'Russia', 'Germany']
sales = [20000, 3000, 5000, 6000, 8000]
plt.bar(country, sales,width=0.2)
plt.show()
```

```python
data_dict = {'India':20000, 'US':3000, 'China':5000, 'Russia':6000, 'Germany':8000}
plt.bar(data_dict.keys(), data_dict.values())
plt.show()
```

# Pie chart

```python
country = ['India', 'US', 'China', 'Russia', 'Germany']
sales = [20000, 3000, 5000, 6000, 8000]
plt.pie(x=sales, labels=country)
plt.show()
```

```python
# showing percentage

India_pct = 20000*100/sum([20000, 3000, 5000, 6000, 8000])
India_pct
```

```python
country = ['India', 'US', 'China', 'Russia', 'Germany']
sales = [20000, 3000, 5000, 6000, 8000]
plt.pie(x=sales,
        labels=country,
        autopct='%.2f%%')  # f is for float value
plt.show()
```

```python
country = ['India', 'US', 'China', 'Russia', 'Germany']
sales = [20000, 3000, 5000, 6000, 8000]
plt.pie(x=sales,
        labels=country,
        autopct='%d%%')  # d is for whole number
plt.show()
```

```python
# using explode
plt.pie(x=sales,
        labels=country,
        autopct='%d%%',
       explode=(0.1, 0.2, 0, 0, 0))  # d is for whole number
plt.show()
```

## Formatting

```python
y = np.arange(1,4)
plt.plot(y, 'y')
plt.plot(y+5, 'm')
plt.plot(y+10, 'c')
plt.show()
```

```python
# control line styling
y = np.arange(1,30)
plt.plot(y,'--', 
        y*4, '-.',
        y*8, ':')
plt.show()
```

## EXtra Topics for Matplotlib

```python
df = pd.read_csv('./global_sales_data/market_fact.csv')
df.head()
```

## box-plot

```python
# box plot
# - visualizes the distribution of a continuous variable
plt.boxplot(df['Order_Quantity'])
plt.show()
```

```python
plt.boxplot(df['Profit'])
plt.show()
```

```python
df['Sales'].describe()
```

```python
# scatter plots are used to visualize two variables , one on each axis
plt.scatter(df['Sales'], df['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
#plt.xlim(0,40000)
plt.show()
```

```python
# enhance the size of the plot
plt.figure(figsize=(10,8)) ########################
plt.scatter(df['Sales'], df['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

# subplotting


```python

# subplot 1: Order_Quantity
plt.subplot(1,2,1)
plt.boxplot(df['Order_Quantity'])
plt.title("Order_Quantity")

plt.subplot(1,2, 2)
plt.boxplot(df['Profit'])
plt.title("Profit")
plt.show()

```
